```python
##Steps in Machine learning
"""
Step 1 - Import libraries
Step 2 - Loading the dataset
Step 3 - Data Preperation
Step 4 - Splitting the dataset(optional)
Step 5 - Model creation
Step 6 - Model training
Step 7 - Model Prediction
"""
```

```python
#Step 1 - Import libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#Step 2 - Loading the dataset
df = pd.read_csv("homeprices.csv")
df.head(10)
```

```python
#Step 3 - Data Preperation
new_df = df.drop('price',axis= 'columns')

X= new_df.values
y= df.price.values

print(type(X))
print(type(y))
```

```python
#Step 5 - Model creation
model = LinearRegression()
```

```python
#Step 6 - Model training
model.fit(X,y)
```

```python
#Step 7 - Model Prediction
print('House price prediction for 2800 area'model.predict([[2800]]))
```

```python
"""
#Linear regression model has 2 predefined attributes -
1. Coefficients - model.coef_
2. Intercept - model.intercept_
"""
print('Coefficient value is : ',model.coef_)
print('Intercept value is : ',model.intercept_)
```

```python
'''
y = mx + c 
m -> Slope
c -> y intercept
'''
#House_price = Coefficient * area + Intercept
House_price = model.coef_ * 2800 + model.intercept_
print(House_price)
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area.values,
           df.price.values,
           color = 'red',
           marker = '*')
plt.plot(df.area.values,
        model.predict(df[['area']].values),
        color = 'blue')
plt.show()
```
